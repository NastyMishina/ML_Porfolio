```python
from functools import reduce
```

```python
class Model(object):
    """   """
    def __init__(self, b_x, b0=0):
        self.b0 = b0
        self.b_x = b_x
        
    def predict(self, x):
        summ = self.b0
        for i in range(0, x.shape[1]):
            summ += self.b_x[i] * x[i] 
        return summ
    
    def error(self, X, Y):    
        return sum((self.predict(X) - Y)**2) / (2 * len(X)) 
    
    def fit(self, X, Y):  
        alpha = 0.1
        for _ in range(500):
            dJ_x = sum(self.predict(X) - Y) /len(X)
            self.b0 -= alpha * dJ_x
            for i in range(0, x.shape[1]):
                dJ_x = sum((self.predict(X) - Y) * X[i]) /len(X)
                new_b = self.b_x[i] - alpha * dJ_x
                self.b_x[i] = new_b
```

```python
regr = Model([1,2,1,1,1,1,1])
```

```python
regr.fit(x, y)
```

```python
yy = regr.predict(x)
plt.scatter(yy, y)
plt.plot(yy, yy, c='r')
```
